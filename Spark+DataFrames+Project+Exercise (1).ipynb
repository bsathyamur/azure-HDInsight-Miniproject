## Bonus Question!
#### There are too many decimal places for mean and stddev in the describe() dataframe. Format the numbers to just show up to two decimal places. Pay careful attention to the datatypes that .describe() returns, we didn't cover how to do this exact formatting, but we covered something very similar. [Check this link for a hint](http://spark.apache.org/docs/latest/api/python/pyspark.sql.html#pyspark.sql.Column.cast)

If you get stuck on this, don't worry, just view the solutions.

```pyspark
walmart_stock.describe().printSchema()
```

```pyspark
from pyspark.sql.functions import col
from pyspark.sql.types import IntegerType

df_desc = walmart_stock.describe()

df_desc_out = df_desc.withColumn("Open",col("Open").cast(IntegerType())) \
              .withColumn("High",col("High").cast(IntegerType())) \
              .withColumn("Low",col("Low").cast(IntegerType()))\
              .withColumn("Close",col("Close").cast(IntegerType())) \
              .withColumn("Volume",col("Volume").cast(IntegerType()))\
              .withColumn("Adj Close",col("Adj Close").cast(IntegerType()))
```

```pyspark
df_desc_out.describe().show()
```

#### Create a new dataframe with a column called HV Ratio that is the ratio of the High Price versus volume of stock traded for a day.

```pyspark
walmart_stock_new = walmart_stock.withColumn('HV Ratio', walmart_stock['High']/walmart_stock['Volume'])
walmart_stock_new.select("HV Ratio").show(20)
```

#### What day had the Peak High in Price?

```pyspark
max_high_value = walmart_stock.agg({"High": "max"}).collect()[0][0]
walmart_stock.filter(walmart_stock.High == max_high_value).select("Date").show()
```

#### What is the mean of the Close column?

```pyspark
round(walmart_stock.agg({"Close": "avg"}).collect()[0][0],2)
```

#### What is the max and min of the Volume column?

```pyspark
max_volume = walmart_stock.agg({"Volume": "max"}).collect()[0][0]
min_volume = walmart_stock.agg({"Volume": "min"}).collect()[0][0]
```

```pyspark
print(max_volume,min_volume)
```

#### How many days was the Close lower than 60 dollars?

```pyspark
walmart_stock.filter(walmart_stock.Close < 60).select("Date").count()
```

#### What percentage of the time was the High greater than 80 dollars ?
#### In other words, (Number of Days High>80)/(Total Days in the dataset)

```pyspark
num1 = walmart_stock.filter(walmart_stock.High > 80).count()
num2 = walmart_stock.count()
percent = (float(num1)/float(num2)) * 100
print(percent)
```

#### What is the Pearson correlation between High and Volume?
#### [Hint](http://spark.apache.org/docs/latest/api/python/pyspark.sql.html#pyspark.sql.DataFrameStatFunctions.corr)

```pyspark
walmart_stock.stat.corr("High","Volume")
```

#### What is the max High per year?

```pyspark
import datetime
from pyspark.sql.functions import year, month, dayofmonth
walmart_stock.groupBy(year("Date").alias("year")).agg({"High": "max"}).sort("year").show()
```

#### What is the average Close for each Calendar Month?
#### In other words, across all the years, what is the average Close price for Jan,Feb, Mar, etc... Your result will have a value for each of these months. 

```pyspark
import pyspark.sql.functions as F
walmart_stock.groupBy(month("Date").alias("month")).agg(F.avg("Close").alias("avg_close")).sort("month").show()
```
